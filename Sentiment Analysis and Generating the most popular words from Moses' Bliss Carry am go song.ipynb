```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Define the text preprocessing function
def preprocess_text(text):
    if isinstance(text, str):
        words = nltk.word_tokenize(text)  # Tokenize the text
        words = [word.lower() for word in words if word.isalpha()]  # Lowercase and keep alphabetic words
        words = [word for word in words if word not in stopwords.words('english')]  # Remove stopwords
        return words
    else:
        return []

# Read the CSV file with the correct delimiter
correct_delimiter = ','  # Replace with the correct delimiter if needed
data = pd.read_csv(r"C:\Users\nanaa\Downloads\Untitled spreadsheet - Moses bliss song.csv", delimiter=correct_delimiter)

# Apply the preprocessing function to the 'Comment' column
data['cleaned_text'] = data['Comment'].apply(preprocess_text)

# Flatten the list of cleaned words and count their occurrences
all_words = [word for words in data['cleaned_text'] for word in words]
word_count = Counter(all_words)

# Display the most common words and their counts
most_common_words = word_count.most_common()
for word, count in most_common_words:
    print(f'{word}: {count}')
```

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Define the text preprocessing function
def preprocess_text(text):
    if isinstance(text, str):
        words = nltk.word_tokenize(text)  # Tokenize the text
        words = [word.lower() for word in words if word.isalpha()]  # Lowercase and keep alphabetic words
        words = [word for word in words if word not in stopwords.words('english')]  # Remove stopwords
        return words
    else:
        return []

# Read the CSV file with pandas
data = pd.read_csv(r"C:\Users\nanaa\Downloads\Untitled spreadsheet - Moses bliss song.csv", delimiter=',')

# Combine all text from the 'Comment' column into a single string
all_text = ' '.join(data['Comment'].dropna().astype(str))

# Preprocess the text
tokens = preprocess_text(all_text)

# Count word frequencies
word_count = Counter(tokens)

# Get the N most common words
N = 20
most_common_words = word_count.most_common(N)

# Unpack words and counts
words, counts = zip(*most_common_words)

# Plot the results
plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top {} Most Common Words".format(N))
plt.xticks(rotation=45)
plt.show()

```

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Define the text preprocessing function
def preprocess_text(text):
    if isinstance(text, str):
        words = nltk.word_tokenize(text)  # Tokenize the text
        words = [word.lower() for word in words if word.isalpha()]  # Lowercase and keep alphabetic words
        words = [word for word in words if word not in stopwords.words('english')]  # Remove stopwords
        return words
    else:
        return []

# Read the CSV file with pandas
data = pd.read_csv(r"C:\Users\nanaa\Downloads\Untitled spreadsheet - Moses bliss song.csv", delimiter=',')

# Combine all text from the 'Comment' column into a single string
all_text = ' '.join(data['Comment'].dropna().astype(str))

# Preprocess the text
tokens = preprocess_text(all_text)

# Count word frequencies
word_count = Counter(tokens)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_count)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Comments")
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download NLTK resources if not already downloaded
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Read the CSV file
csv_file = r"C:\Users\nanaa\Downloads\Untitled spreadsheet - Moses bliss song.csv"
data = pd.read_csv(csv_file)

# Handle NaN values in 'Comment' column
data['Comment'] = data['Comment'].fillna('')  # Replace NaN with empty string

# Convert non-string values to strings
data['Comment'] = data['Comment'].astype(str)

# Function to determine sentiment
def get_sentiment(text):
    sentiment_scores = sid.polarity_scores(text)
    if sentiment_scores['compound'] >= 0.05:
        return "positive"
    elif sentiment_scores['compound'] <= -0.05:
        return "negative"
    else:
        return "neutral"

# Apply sentiment analysis to each comment
data['Sentiment'] = data['Comment'].apply(get_sentiment)

# Output the results
print(data[['Comment', 'Sentiment']])
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Download NLTK resources if not already downloaded
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Read the CSV file
csv_file = r"C:\Users\nanaa\Downloads\Untitled spreadsheet - Moses bliss song.csv"
data = pd.read_csv(csv_file)

# Handle NaN values in 'Comment' column
data['Comment'] = data['Comment'].fillna('')  # Replace NaN with empty string

# Convert non-string values to strings
data['Comment'] = data['Comment'].astype(str)

# Function to determine sentiment
def get_sentiment(text):
    sentiment_scores = sid.polarity_scores(text)
    if sentiment_scores['compound'] >= 0.05:
        return "positive"
    elif sentiment_scores['compound'] <= -0.05:
        return "negative"
    else:
        return "neutral"

# Apply sentiment analysis to each comment
data['Sentiment'] = data['Comment'].apply(get_sentiment)

# Count occurrences of each sentiment category
sentiment_counts = data['Sentiment'].value_counts()

# Plotting the sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Sentiment Analysis')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
